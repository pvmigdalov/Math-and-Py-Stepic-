## Задача №1

### Часть 1

Построить в общих осях графики для:

- $\alpha = 1, \beta = 1$

- $\alpha = 2, \beta = 1$

- $\alpha = 1, \beta = 2$

```python
import numpy as np 
import matplotlib.pyplot as plt

def f(x, alpha=1, beta=1):
    return (x**beta + alpha**beta) / x**beta

x = np.linspace(-5, 5, 100)
x[(x > -0.051) & (x < 0.051)] = np.nan # данный трюк сделан для того, чтобы на графике получить точку разрыва

plt.figure(figsize=(10, 7))
plt.plot(x, f(x, 1, 1), 'r', label=r'$\alpha=1, \beta=1$')
plt.plot(x, f(x, 2, 1), 'g', label=r'$\alpha=2, \beta=1$')
plt.plot(x, f(x, 1, 2), label=r'$\alpha=1, \beta=2$')
plt.ylim(-20, 20)
plt.xlabel(r'$x$', fontsize=14)
plt.ylabel(r'$f(x)$', fontsize=14)
plt.title(r'График функции $f(x)=\frac{x^\beta + \alpha^\beta}{x^\beta}$', fontsize=18)
plt.legend(fontsize=14)
plt.show()

#plt.savefig('graph1.svg')
```

### Часть 2

Построить в общих осях графики для $x>0$.

На том же графике сделать 2 врезки, демонстрирующие поведение графиков на 2 интервалах:

- для малых x

- для больших x

Необходимо продемонстрировать возможность (или невозможность) пересечений и стремление функций.

Цвет линий на врезках и основном графике должен быть одинаковым для одних и тех же значений $\alpha$ и $\beta$.

```python
x = np.linspace(0.001, 1000, 10**5)

fig = plt.figure(figsize=(10, 7))
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.7, 0.65, 0.25, 0.25])
ax3 = fig.add_axes([0.7, 0.25, 0.25, 0.25])


ax1.plot(x, f(x, 1, 1), 'r', label=r'$\alpha=1, \beta=1$')
ax1.plot(x, f(x, 2, 1), 'g', label=r'$\alpha=2, \beta=1$')
ax1.plot(x, f(x, 1, 2), label=r'$\alpha=1, \beta=2$')
ax1.legend(loc='center', fontsize=14)
ax1.set_xlabel('x', fontsize=14)
ax1.set_ylabel('f(x)', fontsize=14)
ax1.set_title('График f(x), x > 0', fontsize=14)
ax1.set_xlim(0, 5)
ax1.set_ylim(0, 100)
ax1.tick_params(axis='both', labelsize=12)

ax2.plot(x, f(x, 1, 1), 'r', label=r'$\alpha=1, \beta=1$')
ax2.plot(x, f(x, 2, 1), 'g', label=r'$\alpha=2, \beta=1$')
ax2.plot(x, f(x, 1, 2), label=r'$\alpha=1, \beta=2$')
ax2.set_xlabel('x', fontsize=14)
ax2.set_ylabel('f(x)', fontsize=14)
ax2.set_title(r'$x \in [0, 0.25]$')
ax2.set_xlim(0, 0.25)
ax2.set_ylim(0, 100)
ax2.tick_params(axis='both', labelsize=12)

ax3.plot(x, f(x, 1, 1), 'r', label=r'$\alpha=1, \beta=1$')
ax3.plot(x, f(x, 2, 1), 'g', label=r'$\alpha=2, \beta=1$')
ax3.plot(x, f(x, 1, 2), label=r'$\alpha=1, \beta=2$')
ax3.set_xlabel('x', fontsize=14)
ax3.set_ylabel('f(x)', fontsize=14)
ax3.set_title(r'$x \in [500, 1000]$')
ax3.set_xlim(500, 1000)
ax3.set_ylim(0.99, 1.015)
ax3.tick_params(axis='both', labelsize=12)

fig.show()

#fig.savefig('graph2.svg')
```

## Задача №2

Построить в общих осях графики для:

- $\alpha = 1, \beta = 0$

- $\alpha = 0, \beta = 0$

- $\alpha = 0, \beta = 1$

- $\alpha = 0, \beta = 2$

- $\alpha = 1, \beta = -1$

- $\alpha = 2, \beta = -1$

Сделайте выводы о поведении графиков

```python
x = np.linspace(-5, 5, 100)
x[(x > -0.051) & (x < 0.051)] = np.nan 

fig, ax = plt.subplots(figsize=(10, 7))
ax.plot(x, f(x, 1, 0), label=r'$\alpha=1, \beta=0$')
ax.plot(x, f(x, 0, 0), label=r'$\alpha=0, \beta=0$')
ax.plot(x, f(x, 0, 1), label=r'$\alpha=0, \beta=1$')
ax.plot(x, f(x, 0, 2), label=r'$\alpha=0, \beta=2$')
ax.plot(x, f(x, 1, -1), label=r'$\alpha=1, \beta=-1$')
ax.plot(x, f(x, 2, -1), label=r'$\alpha=2, \beta=-1$')

ax.legend(loc='best', fontsize=14)
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('f(x)', fontsize=14)
ax.set_title('График f(x)', fontsize=14)

fig.show()

#fig.savefig('graph4.svg')
```

Видим, что если один из коэффициентов равен нулю, то функции возвращают константы. В противном случае наблюдается линейное поведение. 

## Задача №3

Построить в общих осях графики для:

- $\alpha = 1, \beta = 0.5$

- $\alpha = 1, \beta = -0.5$

- $\alpha = 1, \beta = -1.5$

Сделайте выводы о поведении графиков, включая возрастание/убывание и выпуклость/вогнутость

```python
x = np.linspace(-5, 5, 100)
x[(x > -0.051) & (x < 0.051)] = np.nan

try:
    fig, ax = plt.subplots(figsize=(10, 7))
    ax.plot(x, f(x, 1, 0.5), label=r'$\alpha=1, \beta=0.5$')
    ax.plot(x, f(x, 1, -0.5), label=r'$\alpha=1, \beta=-0.5$')
    ax.plot(x, f(x, 0, -1.5), label=r'$\alpha=1, \beta=-1.5$')
except ZeroDivisionError:
    print('При альфа = 0 и бета < 0 функция не определена')
    ax.legend(loc='best', fontsize=14)
    ax.set_xlabel('x', fontsize=14)
    ax.set_ylabel('f(x)', fontsize=14)
    ax.set_title('График f(x)', fontsize=14)

    fig.show()
    #fig.savefig('graph5.svg')
```

При $\alpha = 0$ и $\beta < 0$ функция не определена. В случае $\beta=0.5$ функция убывает и является вогнутой (выпуклой вниз). В другом случае всё с тоности да наоборот. Стоит заметить, что при х < 0 функция не имеет вещественный значений.
